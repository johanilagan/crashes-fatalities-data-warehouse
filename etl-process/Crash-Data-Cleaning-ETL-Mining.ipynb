```python
import pandas as pd
```

```python
crashes = pd.read_csv('bitre_fatal_crashes_dec2024.csv', skiprows = 3, header = 1)
crashes.columns = crashes.columns.str.replace('\n', '', regex = False)
crashes.head(5)
```

```python
fatalities = pd.read_csv('bitre_fatalities_dec2024.csv', skiprows = 3, header = 1)
fatalities.columns = fatalities.columns.str.replace('\n', '', regex = False)
fatalities.head(5)
```

```python
crashes_fatalities = pd.merge(crashes[['Crash ID', 'State','Month','Year','Dayweek','Time','Crash Type','Number Fatalities','Bus Involvement','Heavy Rigid Truck Involvement','Articulated Truck Involvement','Speed Limit','National Remoteness Areas','SA4 Name 2021','National LGA Name 2021','National Road Type','Christmas Period','Easter Period','Day of week','Time of Day']], fatalities[['Crash ID', 'Road User', 'Gender','Age','Age Group']], on='Crash ID', how='inner')

print(crashes_fatalities)
```

```python
# For values that have '-9', it is a missing/unknown value
crashes_fatalities = crashes_fatalities.replace([-9, '-9'], pd.NA)
# For values, specifically in remoteness, that have 'Unknown'
crashes_fatalities = crashes_fatalities.replace('Unknown', pd.NA)
# For values, specifically in road type, that have 'Undetermined'
crashes_fatalities = crashes_fatalities.replace('Undetermined', pd.NA)
# For values that are blank
crashes_fatalities = crashes_fatalities.replace('', pd.NA)
crashes_fatalities.head(10)
```

```python
import pandas as pd
import numpy as np
import mlxtend
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# Getting columns with categorical data types
mining_dataset = crashes_fatalities[[
    "state_name",
    "month_name",
    "day_of_week",
    "crash_type",
    "speed_limit",
    "remoteness_area",
    "road_type",
    "weekday",
    "gender_value",
    "age_grp",
    "road_user_value"]]
mining_dataset

# Checking possible values of the RHS
mining_dataset['road_user_value'].unique()
```

```python
# Converting the dataset into string and dropping NA values
mining_dataset = mining_dataset.astype(str)
mining_dataset = mining_dataset.dropna()

transactions = mining_dataset.values.tolist()

te = TransactionEncoder()
array_te = te.fit(transactions).transform(transactions)

df_encoded = pd.DataFrame(array_te, columns = te.columns_)
df_encoded
```

```python
## Association Rules Mining with mlxtend Library
# Apriori: get frequent itemsets
frequent_itemsets = apriori(df_encoded, min_support=0.1, use_colnames=True)
#Tried min_support = 0.2, but no values applied

frequent_itemsets['length']=frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets['length']

frequent_itemsets[ (frequent_itemsets['length']==2) & 
                  (frequent_itemsets['support']>=0.2)]

#Assume the min confidence is 0.4
rules_con = association_rules(frequent_itemsets, metric="confidence",min_threshold=0.4)
result_arm = rules_con[['antecedents','consequents','support','confidence','lift']]

road_users = ['Driver', 'Passenger', 'Motorcycle rider', 'Pedestrian','Pedal cyclist', 'Motorcycle pillion passenger']
pattern = '|'.join(road_users)

# Filtering the rules where the consequent is a value from the road_user_value
rules = rules_con[rules_con['consequents'].astype(str).str.contains(pattern, case=False, regex=True)]
# Getting the top 5 rules based on lift and confidence
top5_combined = rules.sort_values(by=['lift', 'confidence'], ascending=False).head(5)
top5_combined
```
